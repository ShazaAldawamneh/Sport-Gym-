```python
best_network = load_model("../my_model.h5")
```

```python
path = "../examples"
pathnames = [os.path.join(path, filename) for filename in sorted(os.listdir(path))]

imgs = [load_img(img_path, target_size=(224, 224)) for img_path in pathnames]
for img in imgs:
    plt.figure()
    plt.imshow(img)
    image_array = img_to_array(img)
    image_array = np.expand_dims(image_array, axis=0)
    image_array = preprocess_input(image_array)
    predictions = best_network.predict(image_array)
    print(predictions)
```
